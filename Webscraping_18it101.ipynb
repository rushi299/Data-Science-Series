```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
name=[] 
ratings=[]
genre=[]
time=[]
```

```python
url = "https://www.imdb.com/search/title/?title_type=tv_series&num_votes=100000,&sort=runtime,asc"
content = requests.get(url).content
```

```python
soup = BeautifulSoup(content,"html.parser")

for a in soup.findAll('div', attrs={'class':'lister-item mode-advanced'}):
    h = a.find('h3', attrs={'class':'lister-item-header'})
    a_name=h.find('a', href=True)
    a_rating=a.find('div', attrs={'class':'inline-block ratings-imdb-rating'})
    a_genre=a.find('span', attrs={'class':'genre'})
    a_time=a.find('span', attrs={'class':'runtime'})
    name.append(a_name.text)
    ratings.append(a_rating.text.strip("\n"))
    genre.append(a_genre.text.strip("\n"))
    time.append(a_time.text)


df=pd.DataFrame({'Series Title':name, 'Time':time, 'Rating':ratings, 'Genre':genre})
df.to_csv('/content/web-scraping-18it101.csv', index=False, encoding="utf-8-sig")

df
```
